```python



# Asegúrate de que la columna 'price' esté en formato numérico
calendario['price'] = calendario['price'].replace('[\$,]', '', regex=True).astype(float)

# Calcular el precio medio por 'listing_id'
precio_listing = calendario.groupby('listing_id')['price'].mean().reset_index()

# Renombrar la columna 'price' a 'precio_medio'
precio_listing = precio_listing.rename(columns={'price': 'precio_medio'})



```

```python
df_merged = df_listado.merge(precio_listing, left_on='id', right_on='listing_id', how='left')

# Reemplazar los valores nulos en 'price' con los valores de 'precio_medio'
df_merged['price'] = df_merged.apply(
    lambda row: row['precio_medio'] if pd.isnull(row['price']) else row['price'],
    axis=1
)

# Eliminar las columnas adicionales que no son necesarias después del merge
df_merged = df_merged.drop(columns=['precio_medio', 'listing_id'])
```

```python

# Patrón específico para las licencias:
# xx/Y/ZH/NNNNNN-NN donde:
# xx = 2 números
# Y = 1 letra
# ZH = 2 letras
# NNNNNN = números variables
# -NN = guion seguido de números
license_pattern = r'\d{2}/[A-Z]/[A-Z]{2}/\d+-\d+'

# Identificar licencias válidas
df_merged['is_valid_license'] = df_merged['license'].str.match(license_pattern, na=False)

# Separar en grupos
valid_licenses = df_merged[df_merged['is_valid_license']]['license']
invalid_licenses = df_merged[~df_merged['is_valid_license']]['license']

# Mostrar resultados
print("=== LICENCIAS ENCONTRADAS ===")
print("\nEjemplos de licencias válidas:")
print(valid_licenses.head(10))

print(f"\nTotal de licencias válidas: {len(valid_licenses)}")
print(f"Total de registros no válidos: {len(invalid_licenses)}")

print("\n=== EJEMPLOS NO VÁLIDOS ===")
print(invalid_licenses.head(5))

# Opcional: Mostrar patrones únicos encontrados
print("\n=== PATRONES ÚNICOS ENCONTRADOS ===")
unique_valid = valid_licenses.unique()
print(f"Número de licencias únicas: {len(unique_valid)}")
print("\nEjemplos de licencias únicas:")
print(unique_valid[:5])
```

```python
# Filtrar los datos con 'is_valid_license' igual a True
df_valid = df_limpio[df_limpio['is_valid_license'] == True]

# Importar las librerías necesarias
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión para visualizar la relación
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_valid, x='beds', y='price', hue='neighbourhood_group_cleansed')
plt.title('Relación entre Beds, Price y Neighbourhood Group Cleansed')
plt.xlabel('Número de camas (beds)')
plt.ylabel('Precio')
plt.legend(title='Distrito', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

# Calcular el percentil 75 de 'price' en df_valid
price_75th_percentile = df_valid['price'].quantile(0.75)

# Filtrar los datos para incluir solo aquellos con 'price' por encima del percentil 75
df_high_price = df_valid[df_valid['price'] > price_75th_percentile]

# Crear la gráfica de dispersión con los datos filtrados
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_high_price, x='beds', y='price', hue='neighbourhood_group_cleansed')
plt.title('Relación entre Beds, Price (por encima del percentil 75) y Neighbourhood Group Cleansed')
plt.xlabel('Número de camas (beds)')
plt.ylabel('Precio')
plt.legend(title='Distrito', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Calcular el percentil 999 de 'price' en df_valid
price_999th_percentile = df_valid['price'].quantile(0.999)

# Filtrar los datos para incluir solo aquellos con 'price' por encima del percentil 999
df_high_price = df_valid[df_valid['price'] > price_999th_percentile]

# Crear la gráfica de dispersión con los datos filtrados
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_high_price, x='beds', y='price', hue='neighbourhood_group_cleansed')
plt.title('Relación entre Beds, Price (por encima del percentil 75) y Neighbourhood Group Cleansed')
plt.xlabel('Número de camas (beds)')
plt.ylabel('Precio')
plt.legend(title='Distrito', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
